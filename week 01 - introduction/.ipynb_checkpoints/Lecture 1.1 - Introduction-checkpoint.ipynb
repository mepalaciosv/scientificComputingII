<h1 style="color: teal;">Algorithms</h1>
What is an algorithm?

<img src="https://github.com/mepalaciosv/scientificComputingII/blob/main/images/algorithm.jpg?raw=true" width="400">

Taken from [source](https://www.geeksforgeeks.org/fundamentals-of-algorithms/)

<div style="display: flex; gap: 100px;">

  <div style="flex: 1;">
    <h2 style="color: teal; text-align: center;">Sequential algorithms</h2>
    <ul>
      <li>They only do one operation at a time.</li>
      <li>More common by historical reasons:
        <ul>
          <li>Humans often approach problems sequentially.</li>
          <li>First computers had a single core (processing unit).</li>
        </ul>
      </li>
      <li>There is no need to share memory</li>
    </ul>
  </div>

  <div style="flex: 1;">
    <h2 style="color: teal; text-align: center;">Parallel algorithms</h2>
    <ul>
      <li>Can do multiple operations at a time.</li>
      <li>Became relevant recently (2000s), because
        <ul>
          <li>Multi-core processing units became more popular, and</li>
          <li>Frequency scaling dramatically slowed down due to physical limits.</li>
        </ul>
      </li>
      <li>There is generally a need of sharing memory (communication).</li>
    </ul>
  </div>

</div>

<h2 style="color: teal;">Example</h2>

Calculate the square of the first $n$ natural numbers.

<p><strong style="color: teal;">1. Sequential</strong></p>

```python
import timeit

def square_from_to(limits):
    start, end = limits
    for ii in range(start, end):
        print(ii ** 2, end = " ")

if __name__ == "__main__":
    t = timeit.timeit(lambda: square_from_to((0, 100000)), number = 1)
    print(f"\nExecution time: {t:.6f} s")
```

Complexity: $O(n)$. For $n = 100,000$:

Time spent: 3.340771s

<p><strong style="color: teal;">2. Parallel (with $m$ processing units)</strong> $\rightarrow$ Run this as a .py file, not on Jupyter</p>

```python
import multiprocessing as mp
import timeit

def square_from_to(limits):
    start, end = limits
    for ii in range(start, end):
        print(ii ** 2, end = " ")

def control_unit(n, m):
    with mp.Pool(processes = m) as pool:
        intervals = [(ii, ii + n // m) for ii in range(0, n, n // m)]
        pool.map(square_from_to, intervals)

# This is absolutely needed in Windows, but it could work fine on Linux when removed
if __name__ == "__main__":
    n = 100000
    m = 10
    t = timeit.timeit(lambda: control_unit(n, m), number = 1)
    print(f"Parallel: {t:.6f} s")
```

Complexity: $O(\frac{n}{m})$. For $n = 100,000$:

Time spent: 0.383982s

<h2 style="color: teal;">HPC (High Performance Computing)</h2>

- Scientific simulation and modeling drive the need for greater computer power.
- Single-core processors can not be made fast enough for the simulations (calculations) that are currently needed.
    - Making processors with faster clock speeds is difficult due to cost and power/heat limitations (The **power wall** stopped clock-speed scaling around mid-2000s).
- Solution: **Parallel Computing** - divide the work among numerous linked systems.


<div style="display: flex; align-items: center;">
<div style="flex: 1; text-align: center; margin-right: 30px;">

<img src="https://github.com/mepalaciosv/scientificComputingII/blob/main/images/computer-room-cooling.jpg?raw=true" width="400">

</div>

<div style="flex: 1;">

<h2 style="color: teal;">Generic Parallel Machine</h2>

- Each computer is called a compute **node**.  
- Each node has $m$ processor-cores.  
- These nodes are interconnected by a high-speed network, allowing them to communicate and share workloads. 

</div>
</div>

<h2 style="color: teal;">What do we mean by performance?</h2>

- For scientific and technical programming we use FLOPS (*Floating Point Operations Per Second*).
- Modern Supercomputers reach **PFLOPS** (*PetaFLOPS*) or even **EFLOPS** (*ExaFLOPS*) ranges.
    - Giga ($10^9$), Tera ($10^{12}$), Peta ($10^{15}$), exa ($10^{18}$) FLOPS.

<h2 style="color: teal;">The 80/20 rule</h2>

- Programs typically spend 80% of their time in 20% of the code.
    - This *hot spot* portion is where optimization matters most. 
- Focusing on that critical 20% often gives about 80% of the possible speedup.
    - Getting the remaining 20% of performance usually takes disproportionately more effort. 
- Guideline:
    - Profile before you optimize — find out where the program actually spends time.
    - Do not optimize what is not required (i.e. don’t waste time on code that barely affects runtime).

<h2 style="color: teal;">Metrics of Performance</h2>

<h3 style="color: teal;">What can be measured?</h3>
<ul>
  <li><strong style="color: teal;">Event counts</strong> — how often something happens  
    <ul><li>e.g. cache misses, function calls, page faults</li></ul>
  </li>
  <li><strong style="color: teal;">Time durations</strong> — how long something takes  
    <ul><li>e.g. total run time, function call time</li></ul>
  </li>
  <li><strong style="color: teal;">Resource sizes</strong> — how much of something is used  
    <ul><li>e.g. memory usage, data processed, I/O bytes</li></ul>
  </li>
</ul>

<div style="display: flex; gap: 40px;">

  <div style="flex: 1;">
    <h3 style="color: teal; text-align: center;">Execution Time</h3>
    <ul>
      <li><strong style="color: teal;">Wall-clock time</strong> — total elapsed time as seen by the user</li>
      <li><strong style="color: teal;">CPU time</strong> — total time the CPU spends running your code</li>
    </ul>
    <p style="font-size: 0.9em;">
      <em>Note:</em> CPU time can exceed wall-clock time in parallel programs.
    </p>
  </div>

  <div style="flex: 1;">
    <h3 style="color: teal; text-align: center;">How Are Measurements Taken?</h3>
    <ul>
      <li><strong style="color: teal;">Sampling</strong> — periodically interrupt and record what code is running</li>
      <li><strong style="color: teal;">Instrumentation</strong> — insert timers/counters into the code</li>
    </ul>
  </div>

</div>

<p style="font-size: 0.9em;">
  <em>Note:</em> Measured execution time is non-deterministic, it can vary between runs due to scheduling, caching, and other system activity.
</p>

<h2 style="color: teal;">Typical Performance Analysis Procedure</h2>

<ul>
  <li>
    <h3 style="color: teal;">1. Determine if a performance problem exists</h3>
    <ul>
      <li>Measure baseline performance: execution time, speedup, efficiency, scalability.</li>
      <li>Compare with expected theoretical limits (peak FLOPS, memory bandwidth).</li>
      <li>Ask: Is further optimization worth the effort?</li>
    </ul>
  </li>

  <li>
    <h3 style="color: teal;">2. Identify the type of bottleneck</h3>
    <ul>
      <li>Is the problem <strong>computation-bound</strong> or <strong>communication-bound</strong>?</li>
      <li>Use flat profiling with tools such as <em>MPI</em> or <em>OpenMP</em> profilers to see time distribution.</li>
    </ul>
  </li>

  <li>
    <h3 style="color: teal;">3. Locate the bottleneck in the code</h3>
    <ul>
      <li>Use call-path profiling to see which call chains dominate runtime.</li>
      <li>Apply fine-grained profiling (basic block or line-level) on hot spots.</li>
    </ul>
  </li>

  <li>
    <h3 style="color: teal;">4. Understand why it is slow</h3>
    <ul>
      <li>Use hardware performance counters to detect issues (cache misses, branch mispredicts, vectorization).</li>
      <li>Use tracing on selected regions (keep trace size manageable).</li>
    </ul>
  </li>

  <li>
    <h3 style="color: teal;">5. Evaluate scalability</h3>
    <ul>
      <li>Run at different scales (process/thread counts) and compare profiles.</li>
      <li>Look for load imbalance, communication overhead, or contention.</li>
    </ul>
  </li>
</ul>
